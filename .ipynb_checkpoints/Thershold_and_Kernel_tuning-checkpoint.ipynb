```python
def get_metrics(y_true, y_pred):
    return f1_score(y_true, y_pred), recall_score(y_true, y_pred), precision_score(y_true, y_pred), accuracy_score(y_true, y_pred)

def class_report1(y_true, y_pred, title=''):
    report = classification_report(y_true, y_pred, output_dict= True)
    report_df = pd.DataFrame(report).transpose()
    n = list(report_df.index[:2])
    n.append('accuracy')
    n.append('macro avg')
    n.append('weighted avg')
    report_df.index = n
    display(report_df)
    
    plt.figure(figsize=(8,7))
    #sns.heatmap(confusion_matrix(y_true, y_pred), annot= True, fmt= '.2f')
    color_map = ['crimson', 'lavender'] # Add your desired colors here
    # Create a seaborn heatmap with custom colors
    sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, cmap=sns.color_palette(color_map), fmt = '.2f',
                xticklabels= ['Non Crack', 'Crack'], yticklabels = ['Non Crack', 'Crack'], linewidth=2, linecolor= 'k')
    #plt.xticks([0.5, 1.5], ['Non Crack', 'Crack'])
    #plt.yticks([0.5, 1.5], ['Non Crack', 'Crack'])

    plt.title(title, fontsize= 12)
    plt.xlabel('Predicted')
    plt.ylabel('Truth')
    #plt.show()
    
    print('recall score = {}, f1 score= {}'.format(recall_score(y_true, y_pred), f1_score(y_true, y_pred)))

def Pipeline (X_train, label, image_path, first_iter= True, lab = 'pos', mean_shape = (10,10), diffr = 20, div=2.5):
    '''Returns and appends a preprocessed image into a numpy array that the user creates for the train data'''
    image= cv2.imread(image_path)
    #print(image_path)
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
    gray_image = cv2.resize(gray_image, (512,512))

    # Smooth the image to remove noise
    smooth_image = cv2.blur(gray_image, (3, 3))

    # Calculate the mean and standard deviation of the pixel values using a 3x3 neighborhood
    mean_image = cv2.blur(smooth_image, mean_shape)
    std_image = cv2.absdiff(smooth_image, mean_image)

    # Define a threshold to determine crack regions
    # threshold = 10

    # Automated threshold selection based on histogram distribution in gray image, basically if there's so much brightness in
    # image we want to use a high threshold to activate the crack if not, we use a low threshold

    # Calculating the average of the 25th, 50th, 75th, and 85th percentile pixel value
    perc_25, perc_35, perc_45, perc_50, perc_65, perc_75  = np.percentile(gray_image, 25), np.percentile(gray_image, 35), \
    np.percentile(gray_image, 45), np.percentile(gray_image, 50), np.percentile(gray_image, 65), np.percentile(gray_image, 75)

    # Calculating average of the 1th, 1.2th and 1.5th percentiles
    perc_1, perc_1p2, perc_1p5 = np.percentile(gray_image, 1), np.percentile(gray_image, 1.2), \
    np.percentile(gray_image, 1.5)

    low_mean = np.mean([perc_1, perc_1p2, perc_1p5])
    high_mean = np.mean([perc_25, perc_35, perc_45, perc_50, perc_65, perc_75])
    join_mean = np.mean([perc_1, perc_1p2, perc_1p5, perc_25, perc_35, perc_45, perc_50, perc_65, perc_75])
    
    diff = high_mean - low_mean
    #ratio = low_mean/high_mean
    threshold = join_mean - low_mean

    # For non crack images, we don't expect a high difference so we set a condition
    if diff <= diffr:
        threshold = diffr
    else:
        threshold = np.sqrt(threshold)/div
    # print(threshold)

    # Create a binary mask where cracks are represented as white pixels
    crack_mask = np.zeros_like(std_image)
    crack_mask[std_image > threshold] = 255

    # Apply the crack mask to the original image
    gray_image = gray_image.astype(np.uint8)
    crack_mask = crack_mask.astype(np.uint8)
    segmented_image = cv2.bitwise_and(gray_image, gray_image, mask=crack_mask)
    
    # removing whiter pixels from the segmented image due to the Whitening effect
    segmented_image[segmented_image > join_mean+15] = 0
    
    '''plt.figure(figsize=(8,6))
    plt.subplot(1,2,1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Gray Image')

    plt.subplot(1,2,2)
    plt.imshow(segmented_image, cmap='gray')
    plt.title('Segmented Image')
    plt.grid()
    plt.show()'''
    

    if lab == 'pos':
        label.append(1)
    else:
        label.append(0)

    if first_iter == True:
        row_sum = np.sum(segmented_image, axis=1)
        ar = np.append(np.array([row_sum]), np.sum(segmented_image, axis=0, keepdims=True)).flatten()
        return np.array([ar]), label

    # This implement row wise summation
    # row_sum = np.sum(segmented_image, axis=1)
    # return np.array(row_sum), label
    # This implement col wise summation
    #return np.sum(segmented_image, axis=0, keepdims=True), label
    # This just flattens the entire segmented image
    #return np.array([segmented_image.flatten()]), label
    else:
        row_sum = np.sum(segmented_image, axis=1)
        ar = np.append(np.array([row_sum]), np.sum(segmented_image, axis=0, keepdims=True)).flatten()
        return np.append(X_train, np.array([ar]), axis=0), label
        #return np.append(X_train, np.array([segmented_image.flatten()]), axis = 0), label
        #flattened_image = np.array([segmented_image.flatten()])
        #return np.append(X_train, flattened_image, axis=0), label

def preprocess_data(empty_x, empty_y, cr_data_path, ncr_data_path, data_size = 10000,
                          mean_shape = (16,16), diff = 15, tr_te = 'train', divi = 2):
    # Initializing random seed to enable reproducability 
    random.seed(10)
    
    # tr_te was added because of LDA, you only want to transform on the test set
    cr_content = os.listdir(cr_data_path)[:data_size]
    
    # Randonmly selecting train_data points since I did data Augmentation I didn't want to end up selecting similar
    # Pictures
    cr_content = random.sample(cr_content, k=data_size)
    
    ncr_content = os.listdir(ncr_data_path)[:data_size]
    ncr_content = random.sample(ncr_content, k=data_size)
    

    outputs = Pipeline(empty_x, empty_y, cr_data_path+'/'+cr_content[0], mean_shape = mean_shape, div= divi, diffr=diff)
    empty_x = outputs[0]
    empty_y = outputs[1]

  # adding remaining positive images
    for ind, pics in enumerate(tqdm(cr_content[1:])):
        outputs = Pipeline(empty_x, empty_y, cr_data_path+'/'+cr_content[ind+1], first_iter= False,
                       mean_shape = mean_shape, div= divi, diffr=diff)
        empty_x = outputs[0]
        empty_y = outputs[1]

    # Correctness check
    print('Data size after adding positives is:', empty_x.shape, len(empty_y))

  # adding negative images
    for ind, pics in enumerate(tqdm(ncr_content[:])):
        outputs = Pipeline(empty_x, empty_y, ncr_data_path+'/'+ncr_content[ind], first_iter= False, lab= 'neg',
                       mean_shape = mean_shape, div= divi, diffr=diff)
        empty_x = outputs[0]
        empty_y = outputs[1]

    # Correctness check
    print('\n'+'Data size after adding negatives is:', empty_x.shape, len(empty_y))
    
    # Below, we return the X, y and then the picture names to use in making probability plot and to troubleshoot in cases 
    # of misclassification
    cr_content.extend(ncr_content)
    return empty_x, empty_y, cr_content
```

```python
# REVISE THIS FORMULA
# I revised it, the issue was It doesn't capture widths, like thick cracks, I had to increase the kernel size for mean
# determination because what was happening was that for wide cracks, using a small kernel will fit perfectly in the dark region
# and then the mean - individual pixel values won't produce a large value. Increasing the kernel size worked. 

# Load the image
#image = cv2.imread('image.jpg')
image= cv2.imread('Data/ch/IMG_20230606_110845.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
gray_image = cv2.resize(gray_image, (1024,1024))

# Smooth the image to remove noise
smooth_image = cv2.blur(gray_image, (3, 3))

# Calculate the mean and standard deviation of the pixel values using a 3x3 neighborhood
mean_image = cv2.blur(smooth_image, (5,5))
std_image = cv2.absdiff(smooth_image, mean_image)

# Define a threshold to determine crack regions
threshold = 20

# Create a binary mask where cracks are represented as white pixels
crack_mask = np.zeros_like(std_image)
crack_mask[std_image > threshold] = 255
#crack_mask = crack_mask/255

# Apply the crack mask to the original image
gray_image = gray_image.astype(np.uint8)
crack_mask = crack_mask.astype(np.uint8)
segmented_image = cv2.bitwise_and(gray_image, gray_image, mask=crack_mask)

# Display the segmented image
#cv2.imshow('Segmented Image', segmented_image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

The line segmented_image = cv2.bitwise_and(image, image, mask=crack_mask) applies a bitwise AND operation between the image and itself, using the crack_mask as a mask.

In this line of code, image is the input image or array, and crack_mask is a binary mask that defines the region of interest or the specific pixels to be considered for the bitwise operation.

The cv2.bitwise_and() function is used to perform the element-wise logical AND operation between the corresponding elements of the input arrays (image and image) using the specified mask (crack_mask). The resulting image, segmented_image, will only retain the pixels in image where the corresponding pixel in crack_mask is non-zero (white).

By applying the bitwise AND operation with the mask, the resulting segmented_image will only contain the pixels from image that are within the crack region specified by the crack_mask. This operation effectively segments or isolates the crack region from the rest of the image.

Note that for the cv2.bitwise_and() function to work correctly, the image, image, and crack_mask should have the same size and data type.

## Trying a CNN

```python
def Pipeline_CNN (X_train, label, image_path, lab = 'pos', first_iter = True):
    '''Process an image and returns a numpy array with the image appended with all other images'''

    image= cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Cropping the image
    #image = image[200:1000,625:1260,:]
    # Resizing the image as per Pretrained model Requirement
    image = cv2.resize(image, (224, 224))
    # Changing data type
    image = tf.cast(image, tf.float32)
    # Scaling image pixel values
    image /= 255

    if lab == 'pos':
        label.append(1)
    else:
        label.append(0)

    # Appending just one image to the array on first iteration
    if first_iter == True:
        return np.array([image]), label
    # Appending next image to previous images after first iteration
    else:
        return np.append(X_train, np.array([image]), axis = 0), label


def preprocess_data_CNN(empty_x, empty_y, ret_data_path, not_ret_data_path, data_size = 10000):
  '''Takes an empty 2d array and the retina and not retina image path and returns a numpy array of all images appended'''
  ret_content = os.listdir(ret_data_path)[:data_size]
  nret_content = os.listdir(not_ret_data_path)[:data_size]

  outputs = Pipeline_CNN(empty_x, empty_y, ret_data_path+'/'+ret_content[0])
  empty_x = outputs[0]
  empty_y = outputs[1]

  # adding remaining positive images
  for ind, pics in enumerate(tqdm(ret_content[1:])):
      outputs = Pipeline_CNN(empty_x, empty_y, ret_data_path+'/'+ret_content[ind+1], first_iter= False)
      empty_x = outputs[0]
      empty_y = outputs[1]

    # Correctness check
  print('Data size after adding positives is:', empty_x.shape, len(empty_y))

  # adding negative images
  for ind, pics in enumerate(tqdm(nret_content[:])):
      outputs = Pipeline_CNN(empty_x, empty_y, not_ret_data_path+'/'+nret_content[ind], first_iter= False, lab= 'neg')
      empty_x = outputs[0]
      empty_y = outputs[1]

    # Correctness check
  print('\n'+'Data size after adding negatives is:', empty_x.shape, len(empty_y))
  return empty_x, empty_y
```

```python
# Neural net
image_size= 224

url = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
feature_extract = hub.KerasLayer(url, input_shape = (image_size, image_size, 3))
feature_extract.trainable = False

model = tf.keras.Sequential([
    feature_extract,
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy',
             metrics = ['accuracy'])

# Early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 20)

#saving best model
save_best = tf.keras.callbacks.ModelCheckpoint("./best_model.h5", monitor = 'val_loss',
                                              save_best_only = True)
```

```python
epochs = 10
history = model.fit(x = X_train, y = np.array(label),
                       epochs = epochs,
                       validation_split = 0.15, batch_size = 32,
                       callbacks = [early_stopping, save_best])
```
